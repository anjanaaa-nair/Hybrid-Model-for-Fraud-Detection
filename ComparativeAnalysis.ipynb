```python
import pandas as pd

# Load metrics from CSV files for each model
model1_metrics = pd.read_csv('hb1_metrics.csv')
model2_metrics = pd.read_csv('hb2_metrics.csv')
model3_metrics = pd.read_csv('hb3_metrics.csv')

import matplotlib.pyplot as plt
import numpy as np

# Replace with actual model names
models = ['Hybrid Model 1', 'Hybrid Model 2', 'Hybrid Model 3']
%store models

# Replace with actual accuracy values for each model
accuracy = [0.9996394843747252, 0.9997362080790673, 0.999806552591316]
%store accuracy

# Replace with actual precision values for each model
precision = [0.9996314302262317, 0.9997542738793134, 0.9998946444249341]
%store precision

# Replace with actual recall values for each model
recall = [0.9996489750070205, 0.9997191800056164, 0.9997191011235955]
%store recall

# Replace with actual F1 score values for each model
f1_score = [0.9996402025396436, 0.9997367266344889, 0.9998068650689141]
%store f1_score

# Set the width of the bars
bar_width = 0.2

# Set the positions of the bars on the x-axis
index = np.arange(len(models))

# Create a white-to-blue colormap
colormap = plt.get_cmap('Blues')

# Create subplots for different metrics and set colors using the colormap
plt.figure(figsize=(6, 4))
plt.bar(index, accuracy, bar_width, label='Accuracy', color=colormap(0.5))  # 0.5 represents a middle shade of blue
plt.bar(index + bar_width, precision, bar_width, label='Precision', color=colormap(0.7))  # 0.7 represents a lighter shade of blue
plt.bar(index + 2 * bar_width, recall, bar_width, label='Recall', color=colormap(0.9))  # 0.9 represents an even lighter shade of blue
plt.bar(index + 3 * bar_width, f1_score, bar_width, label='F1 Score', color=colormap(1.0))  # 1.0 represents the lightest shade of blue

# Set x-axis labels and title
plt.xlabel('Models')
plt.ylabel('Metrics')
plt.title('Model Performance Metrics')

# Customize the x-axis labels based on your models
plt.xticks(index + bar_width, models)

# Add a legend to the plot
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
